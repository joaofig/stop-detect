```python
import numpy as np
import pandas as pd
import geopandas as gpd
import sys

def load_data():
    """Loads the CSV files and appends them into a single DataFrame"""
    column_types = {'id': np.int_, 'date': np.string_, 'latitude': np.float_, 'longitude': np.float_}
    
    all_df = dict()
    
    for i in range(10357):
        taxi = i + 1
        try:
            df = pd.read_csv("data/" + str(taxi) + ".txt", dtype=column_types, names=['id', 'date', 'latitude', 'longitude'])
            df.drop_duplicates(inplace=True)
            df['rad_lat'] = np.radians(df['latitude'])
            df['rad_lng'] = np.radians(df['longitude'])
            all_df[str(taxi)] = df
        except:
            print(taxi)
    return all_df
```

```python
taxi_df = load_data()
```
