```python
import numpy as np
import pandas as pd
import geopandas as gpd
import sys

def load_data():
    all_df = dict()
    
    for i in range(10357):
        taxi = str(i + 1)
        try:
            df = pd.read_csv("csv/taxi_" + taxi + ".csv")
            all_df[str(taxi)] = df
        except:
            print(taxi)
    return all_df
```

```python
taxi_df = load_data()
```
